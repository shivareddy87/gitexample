```python
LIMIT = 500
cities = ['New York, NY', 'Newark, NJ', 'Edison, NJ', 'Elizabeth, NJ', 'Secaucus, NJ']
results = {}
for city in cities:
    url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&near={}&limit={}&categoryId={}'.format(
        CLIENT_ID, 
        CLIENT_SECRET, 
        VERSION, 
        city,
        LIMIT,
        "4bf58dd8d48988d10f941735")
    results[city] = requests.get(url).json()
```

```python
results
```

```python
df_venues={}
for city in cities:
    venues = json_normalize(results[city]['response']['groups'][0]['items'])
    df_venues[city] = venues[['venue.id','venue.name', 'venue.location.address','venue.location.city','venue.location.state', 'venue.location.lat', 'venue.location.lng']]
    df_venues[city].columns = ['Id','Name', 'Address','City','State', 'Latt', 'Long']
```

```python
maps = {}
for city in cities:
    city_latt = np.mean([results[city]['response']['geocode']['geometry']['bounds']['ne']['lat'],
                        results[city]['response']['geocode']['geometry']['bounds']['sw']['lat']])
    city_long = np.mean([results[city]['response']['geocode']['geometry']['bounds']['ne']['lng'],
                        results[city]['response']['geocode']['geometry']['bounds']['sw']['lng']])
    maps[city] = folium.Map(location=[city_lat, city_lng], zoom_start=12)

    for lat, lng, label in zip(df_venues[city]['Latt'], df_venues[city]['Long'], df_venues[city]['Name']):
        label = folium.Popup(label, parse_html=True)
        folium.CircleMarker(
            [lat, lng],
            radius=10,
            popup=label,
            color='blue',
            fill=True,
            fill_color='purple',
            fill_opacity=0.8,
            parse_html=False).add_to(maps[city])  
    print(f"Indian Resturants in {city} = ", results[city]['response']['totalResults'])
```

<b>Observation

Upon Observing Secacus NJ and NY City NY Maps , Its is clear that the data is overlapped and most of the resturants are in both sets of data. So its best to drop Secacus to drop from the Analysis

Now Lets find how close these resturants are in every city. By calculating the distance between Mean  coordinates and distance to every resturant in each city we can calcuate the Density.

```python
for city in cities:
    city_lat = np.mean([results[city]['response']['geocode']['geometry']['bounds']['ne']['lat'],
                        results[city]['response']['geocode']['geometry']['bounds']['sw']['lat']])
    city_lng = np.mean([results[city]['response']['geocode']['geometry']['bounds']['ne']['lng'],
                        results[city]['response']['geocode']['geometry']['bounds']['sw']['lng']])
    maps[city] = folium.Map(location=[city_lat, city_lng], zoom_start=11)
    venues_mean_coor = [df_venues[city]['Latt'].mean(), df_venues[city]['Long'].mean()] 
    # add markers to map
    for Latt,Long, label in zip(df_venues[city]['Latt'], df_venues[city]['Long'], df_venues[city]['Name']):
        label = folium.Popup(label, parse_html=True)
        folium.CircleMarker(
            [Latt, Long],
            radius=5,
            popup=label,
            color='blue',
            fill=True,
            fill_color='#3186cc',
            fill_opacity=0.7,
            parse_html=False).add_to(maps[city])
        folium.PolyLine([venues_mean_coor, [Latt, Long]], color="green", weight=1.5, opacity=0.5).add_to(maps[city])
    
    label = folium.Popup("Mean Co-ordinate", parse_html=True)
    folium.CircleMarker(
        venues_mean_coor,
        radius=10,
        popup=label,
        color='green',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(maps[city])

    print(city)
    print("Mean Distance from Mean coordinates")
    print(np.mean(np.apply_along_axis(lambda x: np.linalg.norm(x - venues_mean_coor),1,df_venues[city][['Latt','Long']].values)))
```
